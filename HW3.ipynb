```python
# 1st System
# Given
A=np.array([[3,-1,1],[3,6,2],[3,3,7]])
b=np.array([[1,0,4]]).conj().T
x=np.zeros((3,1))
D=np.diag(np.diagonal(A))
D_inv=np.linalg.inv(np.diag(np.diag(A)))
# create first iteration
x_1=np.dot(np.dot(D_inv,(D-A)),x)+np.dot(D_inv,b)
print('First System')
print('------------')
print('First Iteration')
print(x_1)
# 2nd iteration
x_2=np.dot(np.dot(D_inv,(D-A)),x_1)+np.dot(D_inv,b)
print('Second Iteration')
print(x_2)

# 2nd System
# Given
A=np.array([[2,-2,1,1],[0,-3,0.5,1],[0,0,5,-1],[0,0,0,2]])
b=np.array([[0.8,-6.6,4.5,2]]).conj().T
x=np.zeros((4,1))
D=np.diag(np.diagonal(A))
D_inv=np.linalg.inv(np.diag(np.diag(A)))
# create first iteration
x_1=np.dot(np.dot(D_inv,(D-A)),x)+np.dot(D_inv,b)
print('\n Second System')
print('------------')
print('First Iteration')
print(x_1)
# 2nd iteration
x_2=np.dot(np.dot(D_inv,(D-A)),x_1)+np.dot(D_inv,b)
print('Second Iteration')
print(x_2)
```

## 2. Part B (Gauss Seidel)

## 3. Solving System

```python
def jacobi(A,b,x_i,tol=10**-6,iter=1000,verbose=True):
    import numpy as np
    
    num_iter=0
    D=np.diag(np.diagonal(A))
    D_inv=np.linalg.inv(np.diag(np.diag(A)))
    x_o=x_i-1
    
    while num_iter<iter and np.linalg.norm(x_i-x_o)>tol :
        num_iter, x_o[:] = num_iter+1, x_i
        x_i=np.dot(np.dot(D_inv,(D-A)),x_o)+np.dot(D_inv,b)
        
    # Display Results
    if verbose==True:
        print('\n Jacobi')
        print('===========')
        if num_iter==iter:
            print('x did not converge')
        else:
            print('result')
            print('-------')
            print(x_i)
            print('\n iterations')
            print('-------')
            print(num_iter)
        
    # Return Results
    return x_i, num_iter
    

def gs(A,b,x_i,tol=10**-6,iter=1000,verbose=True):
    import numpy as np
    D=np.diagonal(A)
    x_o=x_i-1
    num_iter=0
    
    while num_iter<iter and np.linalg.norm(x_i-x_o)>tol:
        num_iter, x_o[:] = num_iter+1 , x_i
        for k in range(0,len(x_i)):
            x_i[k]=(b[k]-np.dot(A[k,:k],x_i[0:k])-np.dot(A[k,k+1:],x_i[k+1:]))/D[k]
            
    if verbose==True:
        print('\n Gauss Sidel')
        print('===========')
        if num_iter==iter:
            print('x did not converge')
        else:
            print('result')
            print('-------')
            print(x_i)
            print('\n iterations')
            print('-------')
            print(num_iter)
        
    # Return Results
    return x_i, num_iter


def sor(A,b,x_i,omega=1.1,tol=10**-6,iter=1000,verbose=True):
    import numpy as np
    D=np.diagonal(A)
    x_o=x_i-1
    num_iter=0
    
    while num_iter<iter and np.linalg.norm(x_i-x_o)>tol:
        num_iter, x_o[:] = num_iter+1 , x_i
        for k in range(0,len(x_i)):
            x_i[k]=(1-omega)*x_i[k]+(omega*(b[k]-np.dot(A[k,:k],x_i[0:k])-np.dot(A[k,k+1:],x_i[k+1:]))/D[k])
            #print(x_i)
            
    if verbose==True:
        print('\n Successive Over Relaxation')
        print('===========')
        if num_iter==iter:
            print('x did not converge')
        else:
            print('result')
            print('-------')
            print(x_i)
            print('\n iterations')
            print('-------')
            print(num_iter)
        
    # Return Results
    return x_i, num_iter

# Jacobi
jacobi(A,b,x)
x=np.zeros([n,1])
# Gauss Sidel
gs(A,b,x)
x=np.zeros([n,1])
# Successive Over Relaxation
sor(A,b,x)
```
